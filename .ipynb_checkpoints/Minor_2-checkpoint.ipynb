````python
import mysql.connector

mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="1234",
  database="minor_2"
)
mydb.commit()
my_cursor=mydb.cursor()
````

````python
my_cursor.execute("DROP TABLE IF EXISTS KEYWORD")
sql ='''CREATE TABLE KEYWORD(
   WORD LONGTEXT NOT NULL,
    RANKING INT,
    TITLE LONGTEXT,
    HEADING LONGTEXT,
    LINKS LONGTEXT,
   RESULT FLOAT
)'''
my_cursor.execute(sql)
````

````python
my_cursor.execute("SHOW TABLES")
for x in my_cursor:
    print(x)
````

````python
df[['words','rank','title','h1','links','total_result']] = df[['words','rank','title','h1','links','total_result']].fillna('') # Specific columns
keyword_list=list(df.itertuples(index=False, name=None))
print(keyword_list)
````

````python
insert_keyword="INSERT INTO KEYWORD (WORD,RANKING,TITLE,HEADING,LINKS,RESULT) VALUES(%s,%s,%s,%s,%s,%s)"
my_cursor.executemany(insert_keyword,keyword_list)
my_cursor.execute("SELECT * FROM KEYWORD")
myresult = my_cursor.fetchall()
````

````python
for x in myresult:
      print(x)
````
